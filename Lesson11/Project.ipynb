## Автоматизация оценки и хранение моделей

```python
# храним модели тут, обрабатывает выбраной моделью, возвращает параметры, и точность.
def model_test(model_id, X_train, y_train, X_test, y_test, rf_depth=15):
    
    # Обучение моделей
    if model_id == 1:
        # Простая линейная регрессия, r2 = 0,49
        model = LinearRegression()
        model.fit(X_train, y_train)
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)
        
    if model_id == 2:
        # Простай случайный лес, r2 = 0,73
        model = RandomForestRegressor(n_estimators = 1000, max_depth= rf_depth, random_state = 42, n_jobs= -1 )
        model.fit(X_train, y_train)
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)
        
    if model_id == 3:
        # Lasso
        model = Lasso(random_state=42)
        model.fit(X_train, y_train)
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)

    if model_id == 4:
        model = GradientBoostingRegressor(learning_rate=0.11, max_depth=7, random_state=42)
        model.fit(X_train, y_train)
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)
    
    
    # Общая часть с метрикой
    mse_train = mse(y_train, y_pred_train)
    mse_test = mse(y_test, y_pred_test)
    r2_train = r2(y_train, y_pred_train)
    r2_test = r2(y_test, y_pred_test)
    result = pd.DataFrame( {'Model_id': [model_id],
                            'X_train_count': [X_train.shape[0]],
                            'X_train.columns': [','.join(X_train.columns)],
                            'Last_column': [X_train.columns[-1]],
                            'rf_depth':[rf_depth],
                            'mse_train':[mse_train],
                            'mse_test':[mse_test],
                            'r2_train':[r2_train],
                            'r2_test':[r2_test]}, columns=['Model_id', 'X_train_count', 
                                                           'Last_column', 'rf_depth', 
                                                           'X_train.columns',
                                                           'mse_train', 'mse_test',
                                                           'r2_train', 'r2_test'])
    return result
```

```python
# перебор признаков
def wall_through(X_train, y_train, X_test, y_test, models=0, check=False, rf_depth=15):
    models = [1,2] if models == 0 else models
    runned = 0
    for n in models:
        l = 1 if check == True else len(X_train.columns)
        for m in range(l,len(X_train.columns)+1):
            if runned == 0: 
                result = model_test(n, X_train[X_train.columns[:m]], y_train, X_test[X_test.columns[:m]], y_test, rf_depth=rf_depth)
                runned += 1
                #print(result)
            else:
                result = pd.DataFrame.append(result, model_test(n, X_train[X_train.columns[:m]], y_train, X_test[X_test.columns[:m]], y_test, rf_depth=rf_depth))
                #print(result)
    return result
```
